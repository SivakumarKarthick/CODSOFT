```python
# Seperating features and targets
x=data.drop(columns=['Class'],axis=1)
y=data['Class'].astype(int)          # Converting the data type of the class column
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
# splitting dataset into training and testing datas
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Generating a model with Random Forest Classifier
model=RandomForestClassifier()
```

```python
# Fitting the data in the model
model.fit(x_train,y_train)
```

```python
# Model Evaluation
x_train_prediction=model.predict(x_train)
print(x_train_prediction)
```

```python
# Calculating accuracy of training data
training_accuracy=accuracy_score(y_train,x_train_prediction)
print(training_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
print(x_test_prediction)
```

```python
# Calculating accuracy of testing data
testing_accuracy=accuracy_score(y_test,x_test_prediction)
print(testing_accuracy)
```
